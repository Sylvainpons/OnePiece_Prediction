```python
import pandas as pd
import calendar

# Dictionnaire de correspondance des mois en français vers leur équivalent numérique
mois_mapping = {
    'janvier': '01',
    'février': '02',
    'mars': '03',
    'avril': '04',
    'mai': '05',
    'juin': '06',
    'juillet': '07',
    'août': '08',
    'septembre': '09',
    'octobre': '10',
    'novembre': '11',
    'décembre': '12'
}

# Fonction de conversion de la date
def convertir_date(date_str):
    date_parts = date_str.split()
    if len(date_parts) == 3:
        jour, mois, annee = date_parts
        mois_num = mois_mapping.get(mois.lower())
        if mois_num is not None:
            date_formatted = f"{jour.replace('er', '')}/{mois_num}/{annee}"
            return pd.to_datetime(date_formatted, format='%d/%m/%Y', dayfirst=True)
    return pd.NaT  # Retourne NaT (Not a Time) si la conversion n'est pas possible

# Appliquer la fonction à la colonne "Date de sortie"
OnePiece_df['Date de sortie:'].apply(convertir_date)

# Vérifier les nouvelles valeurs de la colonne "Date de sortie"
print(OnePiece_df['Date de sortie:'])
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from unidecode import unidecode

# Fonction de prétraitement
def preprocess(text):
    # Conversion en minuscules
    text = text.lower()
    # Suppression de la ponctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Suppression des accents
    text = unidecode(text)
    # Tokenisation
    words = word_tokenize(text)
    # Suppression des stopwords
    words = [w for w in words if w not in stopwords.words('french')]
    # Stemmatisation
    stemmer = PorterStemmer()
    words = [stemmer.stem(w) for w in words]
    # Reconstruction du texte
    text = ' '.join(words)
    return text
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Exemple de similarité cosinus entre deux vecteurs de mots
word1 = corpus_embeddings[0]
word2 = corpus_embeddings[1]

similarity_score = cosine_similarity([word1], [word2])
print("Similarité cosinus :", similarity_score)

```

```python
from scipy.spatial.distance import cosine

# Sélectionner les embeddings des chapitres précédents (par exemple, les 10 chapitres précédents)
previous_chapters = corpus_embeddings[-10:]

# Calculer la similarité entre chaque chapitre précédent et le dernier chapitre
similarities = []
for chapter_embedding in previous_chapters:
    similarity = 1 - cosine(chapter_embedding, corpus_embeddings[-1])
    similarities.append(similarity)

# Afficher les similarités pour chaque chapitre précédent
for i, similarity in enumerate(similarities):
    chapter_number = len(corpus_embeddings) - 10 + i + 1
    print(f"Similarité avec le chapitre {chapter_number}: {similarity}")
```

```python
import numpy as np
from sentence_transformers import util

# Liste des personnages
characters = OnePiece['Character'].unique()

# Initialisation des colonnes de changements de similarité des personnages
for character in characters:
    OnePiece[f"Changements_{character}"] = np.nan

# Calcul des changements de similarité des personnages pour chaque chapitre
for i in range(len(OnePiece)):
    chapter_text = OnePiece.loc[i, 'Résumé Rapide_'] + OnePiece.loc[i, 'Résumés_'] + OnePiece.loc[i, 'Notes_'] + OnePiece.loc[i, 'Informations_']
    chapter_embedding = model.encode([chapter_text])

   # Calcul des similarités entre le chapitre courant et les chapitres précédents
    for j in range(i):
        prev_chapter_text = OnePiece.loc[j, 'Résumé Rapide_'] + OnePiece.loc[j, 'Résumés_'] + OnePiece.loc[j, 'Notes_'] + OnePiece.loc[j, 'Informations_']
        prev_chapter_embedding = model.encode([prev_chapter_text])

        for character in characters:
            similarity = util.cos_sim(chapter_embedding, prev_chapter_embedding)
            prev_similarity = OnePiece.loc[j, f"Changements_{character}"]

            if np.isnan(prev_similarity):
                # Si c'est le premier chapitre avec des données pour ce personnage, on utilise la similarité courante
                OnePiece.loc[i, f"Changements_{character}"] = similarity.item()
            else:
                # Sinon, on calcule le changement de similarité par rapport au chapitre précédent
                change = similarity.item() - prev_similarity
                OnePiece.loc[i, f"Changements_{character}"] = change

```
